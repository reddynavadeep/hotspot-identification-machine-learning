```python
def loaddata(delay,arrival,x,y):
    toreturnx=0
    toreturny=0
    for i in range(0,len(x)):
        strp=x[i]
        t=strp.find('-')
        xlower=float(strp[0:t])
        xhigher=float(strp[t+1:len(strp)])
        if(xlower<=arrival and arrival<=xhigher):
            toreturnx=i
            break
    for i in range(0,len(x)):
        strp=y[i]
        t=strp.find('-')
        ylower=int(strp[0:t])
        yhigher=int(strp[t+1:len(strp)])
        if(ylower<=delay and delay<=yhigher):
            toreturny=i
            break
    #print(toreturnx)
    #print(toreturny)
    return toreturnx,toreturny
    
            
            
        
        
```
