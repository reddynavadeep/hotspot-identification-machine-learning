```python
import pandas as pd
import numpy as np
from pandas import ExcelWriter
from pandas import ExcelFile
```

```python
df1=pd.read_excel('delaydatafilter.xlsx')
df1.columns
```

```python
lp=set(df1['from'])
lp=list(lp)
liststation=lp
```

```python
lp1=df1.values.tolist()
lp1
```

```python
matrix=[] #define empty matrix
row=[] #Mistake position 
for i in range(len(liststation)):          # A for loop for row entries 
    a =[] 
    for j in range(len(liststation)):      # A for loop for column entries 
         a.append(0) 
    matrix.append(a)
    ####


```

```python
for i in range(0,len(liststation)):
    fromst=liststation[i]
    for j in range(0,len(liststation)):
        if(matrix[i][j]==0):
            fromto=liststation[j]
            for k in range(0,len(lp1)):
                 if(fromst==lp1[k][1] and fromto==lp1[k][2]):
                     matrix[i][j]=1
                     #matrix[j][i]=1
                     break
```

```python

from pandas import ExcelWriter
from pandas import ExcelFile

towrite1  = pd.DataFrame(columns=['linkid', 'from', 'to',])


k=1

for i in range(0,len(liststation)):
    for j in range(0,len(liststation)):
        if(matrix[i][j]==1):
            towrite1=towrite1.append({'linkid':k ,'from':liststation[i],'to':liststation[j]}, ignore_index=True)
            k=k+1
writer = ExcelWriter('linkdata.xlsx')
towrite1.to_excel(writer,'Sheet1',index=False)
writer.save()
```
